```python
        
from pyspark.sql.types import StructType,StructField,StringType, DoubleType, IntegerType

schema = StructType([StructField("Date", StringType(), nullable=True),
                      StructField("Open", DoubleType(), nullable=True),
                       StructField("High", DoubleType(), nullable=True),
                        StructField("Low", DoubleType(), nullable=True),
                     StructField("Close", DoubleType(), nullable=True),
                     StructField("Volume", IntegerType(), nullable=True),
                     StructField("Adj Close", DoubleType(), nullable=True)
                     ])

# Importing date as String in Schema
```

```python
walmart_df_with_schema = spark.read.format("csv") \
      .option("header","true") \
      .schema(schema) \
      .load("walmart_stock.csv")
```

```python
walmart_df_with_schema.show()
```

```python
# Finally converting the string into date using to_date() function.
from pyspark.sql.functions import *
from pyspark.sql.functions import col, to_date

walmart_df_with_schema = walmart_df_with_schema.withColumn('Date', to_date(col('Date'), 'yyyy-MM-dd'))
walmart_df_with_schema.show()
```
